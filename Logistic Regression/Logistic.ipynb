```python
import pandas  as pd 
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt 
```

Logistic regression formula :  1/ 1+e^x - z(mx+c)

```python
logdf=pd.read_csv("insurance_data.csv")
linedf=pd.read_csv("insurance_data.csv")
plt.scatter(linedf["age"],linedf["bought_insurance"],marker="*",color="red")
```

```python
from sklearn.linear_model import LogisticRegression
x=logdf[["age"]]
y=logdf[["bought_insurance"]]
```

```python
model=LogisticRegression()
model.fit(x,y)
```

# to check the accuracy o logistic regression

```python
answer=model.score(x,y)
print(answer)
```

# importing train and test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33)
```
